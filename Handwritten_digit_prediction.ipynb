```python
import gradio as gr
import tensorflow as tf # mathematical opration performed on neurons
import numpy as np
```

```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
#from keras.datasets import minst
#(x_train, y_train),(x_test,y_test)=tf.keras.datasets.minst.load_data()
```

```python
x_train[0]
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[7],cmap='gray')
plt.show()
```

```python
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
x_train[0]
```

```python
x_train=x_train.reshape(-1,28,28,1)
x_test=x_test.reshape(-1,28,28,1)
#-1:indicates 1 image at a state/ how many image are there in a scale, 28,28: 28x28 dimension image, 1:it used for convert 3D
```

```python
x_train[0]
```

```python
y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)
```

```python
y_train[0]
```

```python
model=tf.keras.Sequential()
model.add(tf.keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(tf.keras.layers.MaxPooling2D(2,2))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='softmax'))
```

sequential is used as a container to store neurons
kernal size 3X3


```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=128,validation_split=0.2)
```

```python
accuracy=model.evaluate(x_test,y_test)
accuracy
```

```python
def sketch_recognition(img):
    x=model.predict(img.argmax(axis=1))
    return x
```

```python
import numpy as np

def classify(image):
  image = np.expand_dims(image, axis=0)
  prediction = model.predict(image).tolist()[0]
  return {str(i):prediction[i] for i in range(10)}

label = gr.outputs.Label(num_top_classes=3)
interface = gr.Interface(classify, 'sketchpad', label, capture_session=True,interpretation="default",title = 'Digit Classifier')
interface.launch(debug=True)
```
