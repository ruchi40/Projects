```python
z=df.drop(columns='Adj Close')
```

```python
x=df.iloc[:, [1,2,3,6]]
x
```

```python
y=df.iloc[:,4]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred2=dt.predict(x_test)
print(f"Predicted close for linear regression {y_pred2} accuracy is {r2_score(y_test,y_pred2)}")


lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred1=lr.predict(x_test)
print(f"Predicted close for linear regression {y_pred1} accuracy is {r2_score(y_test,y_pred1)}")
```
